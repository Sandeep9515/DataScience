```python
import tensorflow as tf
import keras as ks
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
%matplotlib inline
```

```python
fashion_mnist = ks.datasets.fashion_mnist
```

```python
fashion_mnist
```

```python
(train_images , train_labels), (test_images , test_labels) = fashion_mnist.load_data()
# Load data shuffles all the data.
```

```python
randomX, randomY = 1, 2
```

```python
 class_names=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images = train_images / 255.0
test_images = test_images/ 255.0

#0 to 255 is for black and white image
```

```python
plt.figure(figsize=(10,10))
```

```python
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
```

```python
model = ks.Sequential([ks.layers.Flatten(input_shape=(28,28)),ks.layers.Dense(128,activation=tf.nn.relu),ks.layers.Dense(10,activation=tf.nn.softmax)])
```

```python
%%time
model.compile(optimizer=tf.train.AdamOptimizer(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
%%time
model.fit(train_images,train_labels,epochs=100)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```
